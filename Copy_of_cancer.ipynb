```python
num_cols=cancer.select_dtypes(include='number').columns
for col in num_cols:
    Q1 = cancer[col].quantile(0.25)
    Q3 = cancer[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = cancer[(cancer[col] >= lower) & (cancer[col] <= upper)]
```

```python
for col in cancer.columns:
    print(f"\nOutliers in {col}")
    print(cancer.loc[outliers[col],col])
```

```python
outliers_per_column=outliers.sum()
print(outliers_per_column)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="gini",max_depth=3,random_state=42)
```

```python
y_pred = model.predict(X_test)
```
